# Experiment: Collecting Blood with SERs and 5x

It shows that 5x can collect Glucose quite well.

Many Raman Spectra are collected from 5 blood samples using SERs and 5x and 5 accumulations.

```python
from pathlib import Path
from raman import Sample, read_txt, accumulate
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

def add_plot_label():
    plt.xlabel("Raman Shift (cm^-1)")
    plt.ylabel("Intensity")
    plt.grid()

def get_colors(num:int, name:str='cividis') -> np.ndarray:
    cmap = matplotlib.colormaps[name]
    colors = cmap(np.linspace(0, 1, num))
    return colors


DATA_PATH:Path = Path("/app","data")
```

## Each blood sample

There are 

1. blood sample with 88 mg/dL
2. blood sample with 95 mg/dL
3. blood sample with 98 mg/dL
4. blood sample with 121 mg/dL
5. blood sample with 166 mg/dL
6. blood sample with 168 mg/dL

```python
name = "blood-88"
blood_88:list[Sample] = []
colors = get_colors(5)

plt.figure(figsize=(10,8))
for i, path in enumerate(sorted(DATA_PATH.joinpath("bloodSERs","5x","sample").glob(f"{name}_*.txt"))):
    sample = read_txt(path, verbose=True)
    sample.extract_range(low=500, high=2000)
    sample.normalized()
    sample.plot(label=f"{sample.date}", color=colors[i])
    blood_88.append(sample)

plt.vlines(x=1125, ymin=-0.1, ymax=1.1, color='grey', ls='--')
plt.legend()
add_plot_label()
plt.show()
```

The behavior that can be found repeatedly are

1. As the blood degraded/harden, the translucent attribute is reduced. Thus, we can observe the transition of this in the Raman Scattering from the change in the baseline. 
2. As the translucent is vanished, the peak is easier to observed.

Question to ask are

1. Is the Glucose concentration can be observed from the ratio between 1125 and 1450?
2. Is the ratio change according to the degradation?

## Ratio between 1125 and 1450

```python
plt.figure(figsize=(10,8))

colors = get_colors(2, name="PiYG")

for sample in blood_166:
    sample.plot(label="166", color=colors[1])
for sample in blood_95:
    sample.plot(label="95", color=colors[0])

plt.vlines(x=1125, ymin=-0.1, ymax=1.1, color='grey', ls='--')
plt.legend()
add_plot_label()
plt.show()
```

### Smooth the signal

```python
plt.figure(figsize=(10,8))

colors = get_colors(2, name="RdBu")

for sample in blood_95:
    y = sample.smoothing(window_length=30, polyorder=2)
    sample.plot(label="95", color=colors[0])
for sample in blood_166:
    y = sample.smoothing(window_length=30, polyorder=2)
    sample.plot(label="166", color=colors[1])

plt.vlines(x=1125, ymin=-0.1, ymax=1.1, color='grey', ls='--')
plt.vlines(x=1450, ymin=-0.1, ymax=1.1, color='grey', ls='--')
plt.legend()
add_plot_label()
plt.show()
```

```python
DATA_PATH:Path = BASE_PATH.joinpath("data/glucose/5x/txt")

plt.figure(figsize=(16,9))
glucose_5x:list[Sample] = []
for i, path in enumerate(sorted(DATA_PATH.glob("*.txt"))):
    print(i, path)
    sample = read_txt(path=path, verbose=True)
    sample.extract_range(low=600, high=2000)
    sample.normalized(method="minmax")
    glucose_5x.append(sample)
    sample.plot()

plt.title(f"Glucose that collected with 5x lens.")
plt.legend()
add_plot_label()
plt.show()
```

```python
DATA_PATH:Path = BASE_PATH.joinpath("data/glucose/10x/txt")

plt.figure(figsize=(16,9))
glucose_10x:list[Sample] = []
for i, path in enumerate(sorted(DATA_PATH.glob("*.txt"))):
    print(i, path)
    # old name format
    name_format = ["name", "grating", "laser", "exposure", "accumulation", "year", "month", "date", "hour", "minute", "second", "01"]
    sample = read_txt(path=path, verbose=True, name_format=name_format)
    sample.extract_range(low=600, high=2000)
    sample.normalized(method="minmax")
    glucose_10x.append(sample)
    sample.plot()

plt.title(f"Glucose that collected with 10x lens.")
plt.legend()
add_plot_label()
plt.show()
```

```python
glucose_5x_acc = accumulate(glucose_5x)
glucose_10x_acc = accumulate(glucose_10x)
```

```python
DATA_PATH:Path = BASE_PATH.joinpath("data/glucose/10x/txt")

plt.figure(figsize=(16,9))
glucose_5x_acc.plot(label="Glucose - 5x")
glucose_10x_acc.plot(label="Glucose - 10x")
plt.title(f"Compare Glucose that collected with 5x and 10x")
plt.legend()
add_plot_label()
plt.show()
```
